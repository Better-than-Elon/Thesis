```python
import ast
import matplotlib.pyplot as plt
import os
```

```python
def read(path):
    with open(path) as f:
        lines = f.read().rstrip()
    return ast.literal_eval(lines)


def show_plots(path, algo_names, experiment_name, save_folder=None):

    if save_folder is not None and not os.path.exists(save_folder):
       os.makedirs(save_folder)

    plt.figure(3)
    plts = read(path)
    exps_n = len(plts[0])
    algo_n = len(algo_names)
    for exp_id in range(exps_n):
        for grph_id in range(algo_n):
            plt.plot(plts[grph_id][exp_id], label=algo_names[grph_id])
        plt_title = experiment_name + '_Experiment_' + str(exp_id)
        plt.title(plt_title)
        plt.legend()
        if save_folder is not None:
            save_path = save_folder + ('' if save_folder[-1] == '/' else '/') + plt_title + '.png'
            plt.savefig(save_path)
            print(save_path)
        plt.show()
```

```python
algo_names = ['TRPO', 'ATRPO_TD', 'ATRPO_BASIC']
show_plots('data/Human_Trpo_AtpoDelta_Atrpo0_150_0.txt', algo_names, 'Human', 'data/Human_png')
```
